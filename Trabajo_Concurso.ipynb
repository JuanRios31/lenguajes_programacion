```python
# Importar bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

# ID del archivo desde Google Drive
file_id = '1VVydgDkLQ3gRCl3L3KwHuGff04lkOuy1'

# URL de descarga del archivo
url = f'https://drive.google.com/uc?id={file_id}&export=download'

# Leer el archivo CSV desde la URL
df = pd.read_csv(url)

# Verificar columnas
print("Columnas:")
print(df.columns)

# Procesar columna de fechas
df['Date'] = pd.to_datetime(df['Date'])

# Ordenar por fecha
df = df.sort_values('Date').reset_index(drop=True)

# Mostrar las primeras filas
df.head()
```

```python
# 1. Verificar valores nulos por columna
print("Valores nulos por columna:")
print(df.isnull().sum())

# 2. Verificar si hay filas duplicadas
duplicados = df.duplicated().sum()
print(f"\nNúmero de filas duplicadas: {duplicados}")

# Eliminar duplicados si los hay
if duplicados > 0:
    df = df.drop_duplicates().reset_index(drop=True)

# 3. Revisar tipos de datos
print("\nTipos de datos por columna:")
print(df.dtypes)

# Mostrar resumen final
print("\nTamaño final del DataFrame:", df.shape)
df.head()
```

### Rentabilidad Mensual

```python
df['Date'] = pd.to_datetime(df['Date'])

# Usamos 'Date' como índice temporal
df.set_index('Date', inplace=True)

# Calcular rentabilidad mensual: (Precio actual - Precio anterior) / Precio anterior
rentabilidades = df.pct_change()

# Eliminar la primera fila que queda con NaN
rentabilidades = rentabilidades.dropna()

# Opcional: devolver 'Date' como columna si lo prefieres
rentabilidades = rentabilidades.reset_index()

# Mostrar las primeras filas del resultado
rentabilidades.head()
```

```python
# 1. Cálculos sobre rentabilidades mensuales

# Excluimos la columna 'Date' para aplicar funciones a las columnas numéricas (acciones)
rent_mensual_media = rentabilidades.drop(columns='Date').mean()
rent_mensual_min = rentabilidades.drop(columns='Date').min()
rent_mensual_max = rentabilidades.drop(columns='Date').max()

# 2. Rentabilidad acumulada total = ((Precio final / Precio inicial) - 1) * 100
rent_acumulada = ((df.iloc[-1] / df.iloc[0]) - 1) * 100

# 3. Unimos todos los cálculos en un solo DataFrame
resumen = pd.DataFrame({
    'Rentabilidad promedio mensual': rent_mensual_media,
    'Rentabilidad mínima mensual': rent_mensual_min,
    'Rentabilidad máxima mensual': rent_mensual_max,
    'Rentabilidad acumulada (%)': rent_acumulada
})

# Redondear para que se vea mejor
resumen = resumen.round(4)

# Mostrar resumen
resumen.head()
```

### Acción con mayor rentabilidad mensual

```python
accion_mayor_promedio = resumen['Rentabilidad promedio mensual'].idxmax()
valor_mayor_promedio = resumen['Rentabilidad promedio mensual'].max()

print(f"▶️ Acción con mayor rentabilidad promedio mensual: {accion_mayor_promedio} ({valor_mayor_promedio:.4f})")
```

###  Acción que tuvo el mayor valor de rentabilidad mensual positiva en algún mes

```python
# Excluimos la columna 'Date'
rentabilidades_sin_fecha = rentabilidades.drop(columns='Date')

# Encontramos el valor máximo de rentabilidad mensual y la acción correspondiente
valor_max_mensual = rentabilidades_sin_fecha.max().max()
accion_max_mensual = rentabilidades_sin_fecha.max().idxmax()

print(f"▶️ Acción con la mayor rentabilidad mensual positiva: {accion_max_mensual} ({valor_max_mensual:.4f})")
```

### Acción con la mayor rentabilidad acumulada total en el periodo

```python
accion_mayor_acumulada = resumen['Rentabilidad acumulada (%)'].idxmax()
valor_mayor_acumulada = resumen['Rentabilidad acumulada (%)'].max()

print(f"▶️ Acción con mayor rentabilidad acumulada: {accion_mayor_acumulada} ({valor_mayor_acumulada:.2f}%)")
```

### En cuántos meses cada acción tuvo una rentabilidad negativa

```python
# Contamos valores negativos por acción
meses_rent_negativa = (rentabilidades_sin_fecha < 0).sum()

print("▶️ Número de meses con rentabilidad negativa por acción:")
print(meses_rent_negativa.sort_values(ascending=False))
```

### La acción más volátil (mayor desviación estándar mensual)

```python
volatilidad = rentabilidades_sin_fecha.std()
accion_mas_volatil = volatilidad.idxmax()
valor_volatilidad = volatilidad.max()

print(f"▶️ Acción más volátil: {accion_mas_volatil} (desviación estándar: {valor_volatilidad:.4f})")
```

### Acciones que tuvieron más de 70% de meses con rentabilidad positiva

```python
# Porcentaje de meses positivos por acción
meses_totales = rentabilidades_sin_fecha.shape[0]
porcentaje_positivos = (rentabilidades_sin_fecha > 0).sum() / meses_totales

# Filtrar las que tengan más de 70%
acciones_positivas_70 = porcentaje_positivos[porcentaje_positivos > 0.70]

print("▶️ Acciones con más de 70% de meses con rentabilidad positiva:")
print((acciones_positivas_70 * 100).round(2))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

Grafico de linea de 3 acciones (GEB, MINEROS, ECOPETL)

```python
acciones_seleccionadas = ['MINEROS ', 'GEB ', 'ECOPETL ']

# Print the actual column names of the rentabilidades DataFrame
print("Columnas disponibles en rentabilidades:")
print(rentabilidades.columns)

# Gráfico de línea
plt.figure(figsize=(14, 6))
for accion in acciones_seleccionadas:
    # Check if the column exists before plotting
    if accion in rentabilidades.columns:
        plt.plot(rentabilidades['Date'], rentabilidades[accion], label=accion)
    else:
        print(f"Advertencia: La acción '{accion}' no se encontró en las columnas de rentabilidades.")


plt.title("Evolución de la rentabilidad mensual de acciones seleccionadas")
plt.xlabel("Fecha")
plt.ylabel("Rentabilidad mensual")
plt.legend()
plt.grid(True)
plt.show()
```

Grafico de Barras

```python
# Gráfico de barras ordenado por rentabilidad acumulada
resumen_sorted = resumen.sort_values(by='Rentabilidad acumulada (%)', ascending=False)

plt.figure(figsize=(14, 6))
sns.barplot(x=resumen_sorted.index, y='Rentabilidad acumulada (%)', data=resumen_sorted, palette="viridis")
plt.title("Rentabilidad acumulada total (2010–2025)")
plt.xticks(rotation=90)
plt.ylabel("Rentabilidad acumulada (%)")
plt.show()
```

### Boxplot (gráfico de caja): distribución de rentabilidades mensuales por acción

```python
# Convertir rentabilidades a formato largo para boxplot
rentabilidades_long = pd.melt(rentabilidades, id_vars='Date', var_name='Accion', value_name='Rentabilidad')

plt.figure(figsize=(16, 6))
sns.boxplot(x='Accion', y='Rentabilidad', data=rentabilidades_long)
plt.title("Distribución de rentabilidades mensuales por acción")
plt.xticks(rotation=90)
plt.show()
```

Histograma: distribución de todas las rentabilidades mensuales juntas

```python
# Unir todas las rentabilidades en un solo array
todas_rentabilidades = rentabilidades.drop(columns='Date').values.flatten()

plt.figure(figsize=(10, 5))
sns.histplot(todas_rentabilidades, bins=50, kde=True, color='steelblue')
plt.title("Distribución de todas las rentabilidades mensuales (2010–2025)")
plt.xlabel("Rentabilidad mensual")
plt.ylabel("Frecuencia")
plt.show()
```
